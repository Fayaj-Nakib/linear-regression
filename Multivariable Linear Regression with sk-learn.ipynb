# Multivariable Linear Regression with sk-learn

```python
import numpy as np # Numpy Header
import pandas as pd #Pandas Header 
import random #pseudo-random number generators
```

```python
df = pd.read_csv('D:/Linear Regression/seattle-weather.csv') #Inserting Dataset
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder



# Extract the 'weather' column
weather = df['weather']

# Create a LabelEncoder
label_encoder = LabelEncoder()

# Encode the 'weather' column
weather_encoded = label_encoder.fit_transform(weather)

# Replace the original 'weather' column with the encoded values
df['weather'] = weather_encoded

# Now 'weather' is represented as numeric values
print(df.head())
```

```python
print(df.tail()) #Appending Rows
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['precipitation','temp_max','temp_min','wind']].astype(float)
```

```python
Y = df['weather'].astype(float) # output variable (what we are trying to predict)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
# importing sklearn and using linearregression function
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(X, Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

```python
# Predicting the Test set results
y_pred = regr.predict(X_test)

from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print(score)
```

```python
print('Coefficients: \n', regr.coef_)
# The Mean Squared Error
print("Mean squared error: %.2f" % np.mean((regr.predict(X_test) - y_test) ** 2))
# Explained Variance Score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(X_test, y_test))
```
